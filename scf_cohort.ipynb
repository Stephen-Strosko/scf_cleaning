```python
import pandas as pd
data_63 = pd.read_stata('1963_scf.dta')
data_83 = pd.read_stata('1983_scf.dta')
```

## Start Exploration into Individual Cohorts

```python
data_83.head()
```

```python
f = open('scf_smsa.txt', 'r')
smsa_codes = f.read()
```

```python
smsa_codes
```

```python
df = pd.DataFrame({'locations':smsa_codes.split('\n')})
df[['id_number', 'locations']] = df['locations'].str.split('.', 1, expand=True)
df[['metro', 'state']] = df['locations'].str.split(',', 1, expand=True)
df[['state', 'cases_amount']] = df['state'].str.split('(', 1, expand=True)
df[['cases_amount', 'nothing']] = df['cases_amount'].str.split(')', 1, expand=True)
df = df[['id_number', 'metro', 'state', 'cases_amount']]
df
```

```python
data = data_83[['v1', 'b5701', 'b5401', 'b5402']]
data = data.rename({'b5701': 'id_number'}, axis='columns')
```

```python
data.head()
```

```python
data['id_number'] = pd.to_numeric(data['id_number'])
df['id_number'] = pd.to_numeric(df['id_number'])
final_data = pd.merge(data, df, on=['id_number'])
```
